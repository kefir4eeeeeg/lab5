```python
import turtle
import time
```

```python
size = 300
n = 2
turtle.speed(0)

def koch_curve(size, n):
    if n == 0:
        turtle.forward(size)
    else:
        koch_curve(size / 3, n - 1)
        turtle.left(60)
        koch_curve(size / 3, n - 1)
        turtle.right(120)
        koch_curve(size / 3, n - 1)
        turtle.left(60)
        koch_curve(size / 3, n - 1)

start_time = time.time()
koch_curve(size, n)
end_time = time.time() - start_time
print("Время работы: ", end_time)
koch_curve(size, n)
```
